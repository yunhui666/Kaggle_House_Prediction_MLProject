
#### **Write a loop to plot bar plots of each of the categorical features.**


```python
Object = df.dtypes[df.dtypes == 'object'].index
```

```python
df.dtypes[df.dtypes == 'object'].index
```

```python
df.select_dtypes(include = 'object').columns
```

```python
Object
```

```python
# Plot bar plot for each categorical feature
for feature in Object:
    fig = plt.gcf()
    fig.set_size_inches(20, len(df[feature].unique())/ 2)
    
    sns.countplot(y=feature, data=df, order = df[feature].value_counts().index)
    plt.show(feature)
```

## Correlations Matrix
#### the relationships between **numeric features** and **other numeric features**.

```python
# Calculate correlations between numeric features
correlations = df.corr()
```

```python
correlations
```

```python
# Make the figsize 7 x 6
plt.figure(figsize=(7,6))
# Plot heatmap of correlations
sns.heatmap(correlations)
```

```python
# Make the figsize 15 x 14
plt.figure(figsize=(15,14))
# Plot heatmap of correlations
sns.heatmap(correlations)
```

```python
plt.figure(figsize=(16,14))

# Plot heatmap of annotated correlations
sns.heatmap(correlations * 100, annot=True, fmt='.0f')
```

```python
np.zeros_like(correlations, dtype=np.bool)
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
plt.figure(figsize=(12,10))

# Plot heatmap of correlations
sns.heatmap(correlations * 100, annot=True, fmt='.0f', mask=mask)
```

```python
plt.figure(figsize=(16,14))

# Plot heatmap of correlations
sns.heatmap(correlations * 100, annot=True, fmt='.0f', mask=mask, cbar=False)
```
