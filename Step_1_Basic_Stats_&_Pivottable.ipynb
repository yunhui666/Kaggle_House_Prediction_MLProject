## 2. PivotTable 

### I want to know the average lot area for each bldgtype and house style

```python
table = pd.pivot_table(df, values='SalePrice', index=['BldgType'],columns=['OverallQual'], aggfunc=np.mean)
```

```python
table
```

```python
pd.pivot_table(df, values='YearBuilt', index=['BldgType'],columns=['HouseStyle'], aggfunc=np.min)
```

```python
df.groupby(['BldgType','OverallQual'])['SalePrice'].mean()
```

```python
df.groupby(['HouseStyle','BldgType']).mean()
```
